```python
import os
import urllib.request
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np # linear algebra
import matplotlib.pyplot as plt # plotting
import seaborn as sns #sstatistical  data visualisation 
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
categorical_data_train.nunique()
```

```python
from sklearn.compose import ColumnTransformer,make_column_transformer
```

```python
Categorical_oneHotEnc_train=pd.DataFrame([categorical_data_train.Mjob,categorical_data_train.Fjob,categorical_data_train.reason,categorical_data_train.guardian]).T
```

```python
transformer_train = make_column_transformer((OneHotEncoder(sparse=False), ["Mjob","Fjob","reason","guardian"]),remainder='passthrough')
```

```python
transformer_train.get_feature_names_out
```

```python
transformed_data_train = pd.DataFrame(transformer_train.fit_transform(Categorical_oneHotEnc_train),columns=transformer_train.get_feature_names_out())
```

```python
transformed_data_train
```

```python
label_encoder = LabelEncoder()
for col in categorical_data_train :
    if col not in ["Mjob","Fjob","reason","guardian"] :
        categorical_data_train[col] = label_encoder.fit_transform(categorical_data_train[col])
categorical_data_train
```

```python
for c in ["Mjob","Fjob","reason","guardian"] : 
    
    categorical_data_train= categorical_data_train.drop([c],axis = 1)

```

```python
categorical_data_train
```

```python
categorical_data_train= pd.concat([transformed_data_train, categorical_data_train], axis=1)
```

```python
categorical_data_train
```

```python
for c in categorical_data_train :
    print(c,"   ",categorical_data_train[c].unique())
```

```python
from sklearn.preprocessing import StandardScaler , Normalizer
```

```python
scaler = StandardScaler()
```

```python
numerical_data_scaled= pd.DataFrame(scaler.fit_transform(numerical_data_train),columns=[numerical_data_train.columns])
```

```python
numerical_data_scaled
```

```python
for col in numerical_data_scaled:
    print(col)
```

```python
normlize=Normalizer()
```

```python
numerical_data_normlized=pd.DataFrame(normlize.fit_transform(numerical_data_train),columns=numerical_data_train.columns.to_list())
```

```python
numerical_data_normlized.head()
```

```python
final_data_train = pd.concat([numerical_data_normlized, categorical_data_train], axis=1)
final_data_train.head()
```

```python
final_data_train.columns
```

```python
plt.figure(figsize=(70,70))
sns.heatmap(final_data_train.corr(),annot=True)
```
