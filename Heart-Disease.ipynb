## Categorical Data  / Numerical_data

```python
categorical_data=[]
numerical_data=[]
for i,c in enumerate(data.dtypes):
    if c=='object':
        categorical_data.append(data.iloc[:,i])
    else:
        numerical_data.append(data.iloc[:,i])
```

```python
categorical_data=pd.DataFrame(categorical_data).transpose()
numerical_data=pd.DataFrame(numerical_data).transpose()
```

### Relation between categorical data and target

# Data Preprocessing

### Encoding Categorical Data 

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
labelencoder=LabelEncoder()
```

```python
categorical_data.nunique()
```

```python
categorical_data.Diabetic.value_counts()
```

```python
categorical_data.AgeCategory.value_counts()
```

```python
for col in categorical_data:
    categorical_data[col]=labelencoder.fit_transform(categorical_data[col])
```

```python
categorical_data
```

```python
categorical_data.Diabetic.value_counts().sort_index()
```

```python
categorical_data.AgeCategory.value_counts().sort_index()
```

```python
categorical_data.Race.value_counts().sort_index()
```

```python
Categorical_oneHotEnc=pd.DataFrame([categorical_data.Diabetic,categorical_data.AgeCategory,categorical_data.Race])
```

```python
Categorical_oneHotEnc=Categorical_oneHotEnc.T
Categorical_oneHotEnc
```

## OneHotEncoder

```python
from sklearn.compose import ColumnTransformer,make_column_transformer
```

```python
transformer = make_column_transformer((OneHotEncoder(sparse=False), ['AgeCategory', 'Race', 'Diabetic']),remainder='passthrough')
```

```python
transformed_data = pd.DataFrame(transformer.fit_transform(Categorical_oneHotEnc),columns=transformer.get_feature_names_out())
```

```python
transformed_data
```

```python
transformed_data=transformed_data.drop("onehotencoder__AgeCategory_0",axis=1)
```

```python
transformed_data=transformed_data.drop("onehotencoder__Race_0",axis=1)
```

```python
transformed_data=transformed_data.drop("onehotencoder__Diabetic_0",axis=1)
```

```python
transformed_data
```

```python
categorical_data
```

```python
categorical_data = pd.concat([transformed_data, categorical_data], axis=1)
```
